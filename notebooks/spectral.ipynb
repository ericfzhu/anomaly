Import the relevant libraries for this demo

```python
from sklearn.cluster import SpectralClustering
from sklearn.datasets import make_blobs, make_circles
import numpy as np
import matplotlib.pyplot as plt
```

Now we need to prepare a sample dataset to use the `SpectralClustering` method on. Here we will use the `make_blobs` function from the sklearn.datasets submodule, which will generate isotropic Gaussian blobs for clustering.

```python
np.random.seed(2)
X, y = make_blobs(n_samples=400, centers=4, cluster_std=1.5)

plt.scatter(X[:,0], X[:, 1])
plt.show()
```

We can perform spectral clustering on this dataset and visualise the respective generated clusters

```python
cluster = SpectralClustering(n_clusters=3).fit(X)
plt.scatter(X[:,0], X[:,1], c=cluster.labels_)
plt.show()
```

We can also change the number of clusters and see each respective graph

```python
fig, ax = plt.subplots(2, 2)

for x in range(2, 6):
    cluster = SpectralClustering(n_clusters=x).fit(X)
    plt.subplot(2, 2, x - 1)
    plt.scatter(X[:, 0], X[:, 1], c=cluster.labels_, label=f"n_cluster: {x}")
    plt.legend()

plt.show()
```

We can use another sample dataset to better visualise the benefits of using spectral clustering over traditional clustering methods.

```python
X, y = make_circles(n_samples=200, factor=0.5, noise=0.05)

plt.scatter(X[:,0], X[:, 1])
plt.show()
```

```python
cluster = SpectralClustering(n_clusters=2, affinity="nearest_neighbors").fit(X)
plt.scatter(X[:,0], X[:, 1], c=cluster.labels_)
plt.show()
```

And there you have it! Spectral clustering can simply be called from the sklearn library and can be easily applied to any dataset that's suitable for clustering methods.
